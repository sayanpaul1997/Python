#Classification of the Bioactivity

Label: Compounds having values of less than 1000 nM will be considered to be active while those greater than 10,000 nM will be considered to be inactive. As for those values in between 1,000 and 10,000 nM will be referred to as intermediate.

```
classification = []
for i in newdf.standard_value:
    if float(i) >= 10000:
        classification.append("Inactive")
    elif float(i) <= 1000:
        classification.append("Active")
    else: 
        classification.append("intermediate")
```

```
classification
```

Combine the bioactivity classification to pre-processed data

```
dataframe = pd.concat([combine,pd.Series(classification)], axis=1)
```

```
dataframe
```

```
pre_processed_data =dataframe.rename(columns = { 0 : 'Bioactivity_class'}, inplace = False)
```
